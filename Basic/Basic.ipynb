<b>String Formatting<b>

```python
for i in range(1,13):
    print("No. {0} and its squared value is {1} and its cubed value is {2}.".format(i,i**2,i**3))
```

```python
#You can provide any expression inside the format parantheses, they don't always have to variables or literal numbers
```

```python
#In the above example, you can see the numbers are not lined up.
#You can fix it by providing width.
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:4} and its cubed value is {2:4}.".format(i,i**2,i**3))
#Here the number after the colon provide the width
```

```python
#To make the values left aligned, we can do it by providing a < symbol after the colon
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:<3} and its cubed value is {2:<4}".format(i,i**2,i**3))
#Here the values are left aligned
```

```python
#Now to make it centere, we use the ^ symbol
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:4} and its cubed value is {2:^4}".format(i,i**2,i**3))
#Here you will see the all the cubed values are centered
```

```python
#Now you can specify the number of floating point numbers after a decimal with string formatting
```

```python
print("Pi is approx. {0:12}".format(22/7))     #LINE 1
print("Pi is approx. {0:12f}".format(22/7))    #LINE 2
print("Pi is approx. {0:12.50f}".format(22/7)) #LINE 3
print("Pi is approx. {0:52.50f}".format(22/7)) #LINE 4
print("Pi is approx. {0:62.50f}".format(22/7)) #LINE 5
print("Pi is approx. {0:72.50f}".format(22/7)) #LINE 6
```

```python
#In Line 1, it has a width of 12, and a default floating point value to 15
#In Line 2, it has a width of 12, but here we define a floating value "f", it will print default to 6 decimals
#In Line 3, The width 12 has been overridden because precision is more important than width
#In Line 4,5,6 are just the same number with different widths
#NOTE: The maximum decimal floating point number in Python is betwee 52 and 53. 
```

```python
print("Pi is approx. {0:<72.50f}".format(22/7))
print("Pi is approx. {0:<72.54f}".format(22/7))
```

```python
#As ypu can see, we have left aligned the deciaml numbers, in the second line even increasing te float value to 54 we
#only get a single digit at the end, and the rest of them are padded with zeros.
```

```python
#Also to mention that, if you dont specify the indexes in the string in the .format medthod, it will print sequentially,
#and you can even use width.
```

- CONTINUE and BREAK statements

```python
"""Let's see an algo"""
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i != "phone":
        print(f"Buy {i}")
```

```python
#This is one way to exclude the item phone, but the other way is to exclude is to use the oondition continue
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i == "phone":
        continue
    else:
        print("Buy "+i)
```

```python
#now let's see the break condition
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i == "phone":
        break
    else:
        print("Buy "+i)
#here the break statement will break the loop and exit out of the whole loop, any code after that will not be printed
```

```python
#the break condition is generally useful when you need to look through thousands of item for a single item,
#the moment you find it, you would want to exit out of the loop and not execute the rest of the program.
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

found_at=None #to declare this variable dosen't contain anything; None is a constant that represent nothing
for i in range(len(shopping)):
    if shopping[i]=="phone":
        found_at=i
        break
print(found_at)
```

```python
available_parts = ["computer",
                   "monitor",
                   "keyboard",
                   "mouse",
                   "mouse mat",
                   "hdmi cable",
                   "dvd drive"
                   ]
valid_choices=[]
for i in range(1, len(available_parts)+1):
    valid_choices.append(str(i))
print(valid_choices)
current_choice = "-"
computer_parts = []

while current_choice != '0':
    if current_choice in valid_choices: 
        index=int(current_choice)-1
        chosen_part = available_parts[index]
        if chosen_part in computer_parts:
            #if the item is already in the list, so remove it
            print("Removing {}".format(current_choice))
            computer_parts.remove(chosen_part)
        else:
            print("Adding {}".format(current_choice))
            computer_parts.append(chosen_part)
        print(f"Your list now contains: {computer_parts}")
    else:
        print("Please add options from the list below:")
        for number,part in enumerate(available_parts):
            print("{0}: {1}".format(number + 1, part))

    current_choice = input()

print(computer_parts)
```

### Removing items from a list Backwards

```python
#Removing unwanted items from a unsorted list, can be done much more easier than removing unwanted items foreward.
```

```python
#We have seen that we cannot remove items from a list, while iterating over it foreward. We remove an item, the size of 
#the list changes, and will not work properly
```

```python
#Which basically messes up the index numbers as we move towards the list.
```

```python
#But if we were to remove an item from the list, while iterating over it backwards, all the items index after it will be
#changed. But the items indexes before it, will stay exactly the same.
```

```python
#Iterating backwards is a valuable techinque & allows the size of your sequence to be changed without causing a problem.
```

##### Lets see two ways to do that

```python
#Using for loop to iterate backwards
```

```python
data=[104,101,4,105,308,105,5,107,100,306,106,102,108]
min_valid=100
max_valid=200

for index in range(len(data)-1,-1,-1):
    if data[index] < min_valid or data[index] > max_valid:
        print(index,data)
        del data[index]

print(data)
```

```python
#Using Reserved Function
```

```python
#The advantage of using reversed is that it let us use enumerate
```

```python
data=[104,101,4,105,308,105,5,107,100,306,106,102,108]
min_valid=100
max_valid=200

top_index=len(data)-1 #The reason we are using this because we are displaying the index position of the item we are
                      #deleting, you can ignore this, if you don't want to display all index values and just want the
                      #final list.
for index,value in enumerate(reversed(data)):
    if value<min_valid or value>max_valid:
        print(top_index - index,value)
        del data[top_index - index]

print(data)
```

```python
#The 'end' argument
```

```python
#for end to see in action, we will use the code from the previous section
```

```python
menu=[
    ["eggs","bacon"],
    ["eggs","sausage","bacon"],
    ["eggs","spam"],
    ["eggs","bacon","spam"],
    ["eggs","bacon","sausage","spam"],
    ["spam","bacon","sausage","spam"],
    ["spam","eggs","spam","spam","bacon","spam"],
    ["spam","bacon","spam","sausage","spam","tomato"],
]

for meal in menu:
    for item in meal:
        if item !="spam":
            print(item, end=" ") #We have added end argument here
    print()
```

```python
# What this end argument does, instead of printing a new line after each item, they will now be seperated by " "
```

```python
#The print() in line 16 is needed, otherwise every item will be in the same line
```

```python
# Alright. Now we can't use sep here,
# because we've only passed one object to print - you can see that on line 15.
# Remember that the separator's only used between the values we print, when we pass
# more than one value to the print function. However, we can still get a
# comma between the meal items, by putting a comma in the string that we used for
# end, so let's do that.
```

```python
for meal in menu:
    for item in meal:
        if item !="spam":
            print(item, end=", ") #We have added end argument here
    print()
```

```python
# That's not bad, but did you expect that extra comma at the end of each line? It's not
# actually easy to remove that, with this code. In fact, the most pythonic way to
# get that output without a trailing comma, is to use a generator expression.
```

```python
# (Now generator expressions are advanced Python expression, we will learn about it later)
```

```python
# Let's see the generator code, that will remove the comma from the end
```

```python
for meal in menu:
    items = ', '.join((item for item in meal if item != "spam"))
    print(items)
```

```python
# In this coding exercise, you have to write a function named "sum_eo" with the following exercise:
# n: a positive number
# t: a str
# If 't' has a value 'e', the function should return the sum of all 'even' natural numbers less than 'n'.
# Else if 't' has the value'o', the function should return the sum of all 'odd' natural numbers less than 'n'.
# For any other value of 't' return -1.
# Examples:
# sum_eo(10,'e') : should return 20, since 2+4+6+8=20
# sum_eo(7,'o') : should return 9, since 1+3+5=9
# sum_eo(11, 'spam') : should return -1
```

```python
def sum_even(n):
    num=0
    for i in range(n):
        if i%2==0:
            num+=i
    return(num)
    
    
def sum_odd(n):
    num=0
    for i in range(n):
        if i%2!=0:
            num+=i
    return(num)


def sum_eo(n,t):
    if t=='e':
        return sum_even(n)
    elif t=='o':
        return sum_odd(n)
    else:
        return(-1)
    
    
n=int(input("Enter a positive number: "))
t=input("Enter 'e' for even, and 'o' for odd: ")
output=sum_eo(n,t)
print(output)
```

```python
def factorial(num:int) -> num:
    """
    It returns the factorial of a number.
    
    :param num: The number that is being input by the user.
    :return: It will be the factorial of thr inout number.
    """
    j=1
    if num==0 or num==1:
        return 1
    else:
        for i in range(2,num+1):
            j=j*i
        return j


NUM=int(input("Enter a number: "))
print(factorial(NUM))
```

```python
# Lets see what does the *(star) means, and why it is infornt of a variables.
# When you use an * before an object, it unpacks the sequence.
# Lets see an example:

numbers=(1,2,3,4,5,6)

print(numbers, sep=";") #Here we print the tuple
print(*numbers, sep=";") #Here we unpack the sequence and then print the unpacked numbers that the print fucntion got
print(1,2,3,4,5,6, sep=";") #Something like this. #When you run this, youll see line 8 and line 9 are exactly the same.
```

```python
# Notice that we get ';' on the last two, but in the first one, we didnt get one, as they are already separated by ','
```

```python
# You can use * to unpack any sequence type.
# Its very useful sometimes, but it isnt something that you'll do very often.
# Infact you may never do it.
# But, where it gets incredibly useful is when we use it the other way around.
# Instead of unpacking the sequence that we pass as an argument, we can specify that a parameter is an unpacked sequence
```

```python
# Let's see that in action.
```

```python
def test_star(*args):
    print(args)
    for x in args:
        print(x)

test_star(1,2,3,4,5,6)
```

```python
# Inside our function, args is tuple. We can see that on the first line of the output.
# The function then iterates over args and then prints each of its items.
# So, on line 1 in our function definition, the * means this paramter will be replaced by an unpacked tuple.
# When the function's called, the values will be unpacked up into a tuple, 
# and we can see that on line 6 when we call the function.
# Using *args means that we can provide a variable number of arguments to our functions.
# On line 6, we pass 6 arguments, we can pass more arguments or no arguments.
```

```python
def test_star(*args):
    print(args)
    for x in args:
        print(x)

test_star()
```

```python
# Her we get a empty tuple as an output.
# But more importantly, our function didn't crash.
# So, we now have a way to write a function that can take 0 or more arguments, just like the built-in print function.
```

### .join

```python
# Now you know, strings are immutable.
# As a result, it is not efficient to concatenate strings in a loop.
# Everytime you concatenate a string to an existing string, the new string has to be created and
# if it's done in a loop, it become expensive and slow.
# So you wanna avoid that.
# So the string ".join" method is used instead.
# It actually takes a sequence and produces a string from it.
# So, the items in the new string are separated by the string that 'join' was called upon.
```

```python
# so lets see an example:

myList = ["a","b","c","d"]
          
newString=''
for i in myList:
    newString += i + ", "

print(newString)
```

```python
# This is highly inefficeint code. There are number of problems here.
# Firstly, there is comma at the end of the output.
# Secondly, every time through the loop it creating a new copy of the new string(bcuz the string we are using immutable)
# Now here you will use .join to fix the problem.
```

```python
# lets see how to do that

myList = ["a","b","c","d"]
          
newString=", ".join(myList)

print(newString)
```

```python
# So this works, bcuz it creates a new string by taking each item in a sequence(which is a list, in this case)
# And separates the items with a string comma that join is actually invoked on.
# So we can any sequence item as a paramter to join(e.h instead of list we can use a another string)

letters="abcedfghijklmnopqrstuvwxyz"

newString = ", ".join(letters)
print(newString)
```

```python
numbers = "123456789"

newString = " mississipi, ".join(numbers)
print(newString)
```

```python
# Now sets can be subtracted and which way you use is really just a matter of preference.
# Now, we'll cheat a little bit before we check that because all though Python's
# quite happily dealing with unordered lists,
# us humans find them hard to read when of course there's more than a few members.
# So to make it easier to see what's going on, what we're gonna do is use the sorted
# function to sort the results before printing them out.
# Now this is not necessary when dealing with sets. I just wanna make that clear.
# We're only doing it here to make it easier for us to see
# what the various methods are actually doing with the actual sets themselves.
```

```python
# So subtracting set b from set a removes any item that exists in set b from set a
```

```python
even=set(range(0,40,2))
print(sorted(even))
squares_tuple = (4,6,9,16,25)
squares = set(squares_tuple)
print(sorted(squares))

print("Even minus squares")
print(sorted(even.difference(squares)))
print(sorted(even-squares))

print('Squares minus even')
print(sorted(squares.difference(even)))
print(sorted(squares-even))
```

```python
# As yoou can see, using the .difference method or using the minus(-) results in the same exact output.
# So, it really just a matter of prefernce.
# But you should use the .difference method as it helps you, while reading the code that you are working with sets
# but if you use the minus(-) it creates confusion that you may just be subtracting two variables
```

```python
# There is also difference_update method, and that performs a subtraction in place.
# So it dosen't return a new set, what it does is it modifies the set upon which is called on,
# as a result it returns None
```

```python
even = set(range(0,40,2))
print(sorted(even))
squares_tuple = (4,6,9,16,25)
squares = set(squares_tuple)
print(sorted(squares))
even.difference_update(squares)
print(sorted(even))
```

```python
# As you can see, all the items from the squars set is removed from the even set, when we used difference_update method
```

```python
# SYMMETRIC DIFFERENCE OF TWO SETS 
```

```python
# It means that: all the members that are in one set or in the other but not both.
# So being symmetric also dosen't matter which way round you calculate the difference.

# And we will sort out the result before printing it, because it is easier to interpret, you dont need to do that 
# while you are coding in general or working with sets
```

```python
even = set(range(0,40,2))
print(sorted(even))
squares_tuple = (4,6,9,16,25)
squares = set(squares_tuple)
print(sorted(squares))

print("Symmetric even minus squares:")
print(sorted(even.symmetric_difference(squares)))

print("Symmetric squares minus even:")
print(sorted(squares.symmetric_difference(even)))
```

```python
# As you can see, they are same, it dosent matter which way you use it.
```

```python
even = set(range(0,40,2))
print(even)
squares_tuple = (4,6,9,16,25)
squares = set(squares_tuple)
print(squares)

squares.discard(4)
squares.remove(16)
squares.discard(8)  # NO ERROR, (8 is not a part of 'squares')
print(squares)
squares.remove(8)   # ERROR
```

```python
# These are used to test if one set is a sub or superset of the other set.
# Now one set is a subset of anbother if all the members are contained in the other set.
# And similarly, One set is a superset of another if it contains all the other sets and members. 
```

```python
even = set(range(0,40,2))
print(even)
squares_tuple = (4,6,16)
squares = set(squares_tuple)
print(squares)

if squares.issubset(even):
    print("Squares is a subset of even")
if even.issuperset(squares):
    print("Even is a superset of squares")
```

```python
# There is a more Pythonic way to open and also removes the risk of closing the files afterwards & it is more concise.
# So, Python has a 'with' statement, and if a object supports the use of 'with' and a file does, the 'with' will tidy
# everything up automatically once the object is no longer needed
```

```python
with open(r"C:\\Users\\Kaustubh\\Documents\\Python\\Basic\\sample.txt") as jabber:
    for line in jabber:
        if "jabberwock" in line.upper():
            print(line, end = '')
```

```python
# Now we dont need to explicitly close the file, the with takes cares of it for us.
# You'll also notice that these open files are associated with a variable by using 'as jabber'
# So that we can refer to jabber in the loop.
# There is also an advantage of using with, if a error occurs while handling the file, earlier it would have stopped
# but using with, it handles the files in itself, and traps the error or exception and closes the file before the
# error can terminates the program.
```

```python
with open(r"C:\\Users\\Kaustubh\\Documents\\Python\\Basic\\sample.txt") as jabber:
    line = jabber.readline()
    while line:
        print(line, end = '')
        line = jabber.readline()
```

```python
# Here you can see we get the whole poem this time, without any double spacing because
# end ='' has removed that extra line, because of course there was an end line in the actual text file itself.
# And so, what we're saying is, by putting end='' or an empty string, we're saying don't print that line as well,
# and thats why we are not getting the double up anymore.
# And in term of using readline(), you can see we read the first line of the file on line 2, into the variable line.
# and if some data can be read, in other words, the file wasn't empty, then we enter the while loop.
# Now if there was a problem, and the file was empty, it wont process from line 3 onwards.
# And at the point the while loop is gonna contiune until there's no line.

# And this code is probably closer to the way that a file is normally processed in other programming languages,
# sort of line by line as you can see there.
```

```python
# So the process of writing data to a text file is very similar to reading from it.
# The files opened but this time the mode is w from write instead of r for read
# the actual data is written using the print function exactly as you want to display it on the screen but
# only this time we specify a file object to send it to and it goes to the file.
# And importantly close the file when you are finished
```

```python
cities = ["Adelaide","Alice Springs","Darwin","Melbourne","Sydney"]

with open("cities.txt",'w') as city_file:
    for city in cities:
        print(city, file=city_file) #instead of just having to print by itself we're adding the file part of the line
                                    #which indicates that we want to write this to a text file.
```

```python
# When you specify the write mode the file will be created if it does not exist, but importantly if it does exist
# it will be overwritten
```

```python
# print in line 5 also has a flush argument that can be set to true or false,
# and the data is false, so data isnt flushed.
# Now to understand what it is for we have to look a little bit detail about whats happening.
# When Python writes data to a file.
# Now, external devices on your computer including the screen are much slower than computer memory.
# So, running with things like disk drives in the screen is buffered.
# What that means is that the data is actually written to a buffer and the contents of that buffer is transferred to the
# external device in the background.
# So it's something that happens, While you're sort of doing something else.
# This allows the program to continue processing without waiting for the write to complete.
# And so, everything seems much faster.
# So in other words, as soon as we run this program, Once the process finished,
# at that point it may still be writing something in the background to that file.
```

```python
# Now this, obviously an example,you have five words that would have been pretty quick.
# But you can imagine like a huge file writing that was perhaps many megabytes long.
# It does take the computer some time to process.
# And that's because it is putting in that buffer It all the
# comes down to the speed of the device that you're writing the information to.So
# essentially it means that the data is actually written to the buffer and then
# the contents of the buffer is transferred to the external device in the background.
```

```python
# ometimes though with that said, you want the data to be written out immediately.
# So especially if the external device is a screen, and
# you want the user to be able to see the output straight away.
# So with buffering, data could be sent to the screen from the buffer for
# example, and then immediately overwritten by subsequent data from the buffer.
# So this wouldn't be apparent in our example so far. Because it already using relatively small amounts of data but
# displaying the last on the screen could suffer from that effect.
```

```python
# As the scrolls up the screen lines of text appear to flicker.
# Now, closing a file causes the buffer to be flushed automatically.
# But if you wanna ensure your data's written sooner,
# you can pass flush equals true to cause the data to be written immediately.
# Generally though with that said, it's not something you'd, you really should worry
# about, but the ability to control it through a window is important
# enough that the benefits Python introduced the flush parameter in Python 3.
```

```python
#Now we get the output we wanted but \n does not look good, so lets remove it as we are importing the file.

cities=[]

with open(r"C:\\Users\\Kaustubh\\Documents\\Python\\Basic\\cities.txt") as city_file:
    for city in city_file:
        cities.append(city.strip("\n")) # use strip to remove the \n escape characters

print(cities)
for city in cities:
    print(city)
```

```python
# So the contents of the file probably look like a good representation of our tuple,
# but there's no easy way to read it back into a tuple variable,
# because it's now stored as a string.
# So, Python does have an eval function that will evaluate any expression
# to a string passed to it, so we could get the contents back from the file into a tuple variable by using eval.
```

```python
with open(r"C:\\Users\\Kaustubh\\Documents\\Python\\Basic\\eminemsongs.txt") as eminem_file:
    contents = eminem_file.readline()

eminem = eval(contents)

print(eminem)
title,artist,year,tracks = eminem
print(title)
print(artist)
print(year)
```
