<b>String Formatting<b>

```python
for i in range(1,13):
    print("No. {0} and its squared value is {1} and its cubed value is {2}.".format(i,i**2,i**3))
```

```python
#You can provide any expression inside the format parantheses, they don't always have to variables or literal numbers
```

```python
#In the above example, you can see the numbers are not lined up.
#You can fix it by providing width.
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:4} and its cubed value is {2:4}.".format(i,i**2,i**3))
#Here the number after the colon provide the width
```

```python
#To make the values left aligned, we can do it by providing a < symbol after the colon
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:<3} and its cubed value is {2:<4}".format(i,i**2,i**3))
#Here the values are left aligned
```

```python
#Now to make it centere, we use the ^ symbol
```

```python
for i in range(1,13):
    print("No. {0:2} and its squared value is {1:4} and its cubed value is {2:^4}".format(i,i**2,i**3))
#Here you will see the all the cubed values are centered
```

```python
#Now you can specify the number of floating point numbers after a decimal with string formatting
```

```python
print("Pi is approx. {0:12}".format(22/7))     #LINE 1
print("Pi is approx. {0:12f}".format(22/7))    #LINE 2
print("Pi is approx. {0:12.50f}".format(22/7)) #LINE 3
print("Pi is approx. {0:52.50f}".format(22/7)) #LINE 4
print("Pi is approx. {0:62.50f}".format(22/7)) #LINE 5
print("Pi is approx. {0:72.50f}".format(22/7)) #LINE 6
```

```python
#In Line 1, it has a width of 12, and a default floating point value to 15
#In Line 2, it has a width of 12, but here we define a floating value "f", it will print default to 6 decimals
#In Line 3, The width 12 has been overridden because precision is more important than width
#In Line 4,5,6 are just the same number with different widths
#NOTE: The maximum decimal floating point number in Python is betwee 52 and 53. 
```

```python
print("Pi is approx. {0:<72.50f}".format(22/7))
print("Pi is approx. {0:<72.54f}".format(22/7))
```

```python
#As ypu can see, we have left aligned the deciaml numbers, in the second line even increasing te float value to 54 we
#only get a single digit at the end, and the rest of them are padded with zeros.
```

```python
#Also to mention that, if you dont specify the indexes in the string in the .format medthod, it will print sequentially,
#and you can even use width.
```

- CONTINUE and BREAK statements

```python
"""Let's see an algo"""
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i != "phone":
        print(f"Buy {i}")
```

```python
#This is one way to exclude the item phone, but the other way is to exclude is to use the oondition continue
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i == "phone":
        continue
    else:
        print("Buy "+i)
```

```python
#now let's see the break condition
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

for i in shopping:
    if i == "phone":
        break
    else:
        print("Buy "+i)
#here the break statement will break the loop and exit out of the whole loop, any code after that will not be printed
```

```python
#the break condition is generally useful when you need to look through thousands of item for a single item,
#the moment you find it, you would want to exit out of the loop and not execute the rest of the program.
```

```python
shopping=["bread","butter","cheese","phone","tea","sugar"]

found_at=None #to declare this variable dosen't contain anything; None is a constant that represent nothing
for i in range(len(shopping)):
    if shopping[i]=="phone":
        found_at=i
        break
print(found_at)
```

```python
available_parts = ["computer",
                   "monitor",
                   "keyboard",
                   "mouse",
                   "mouse mat",
                   "hdmi cable",
                   "dvd drive"
                   ]
valid_choices=[]
for i in range(1, len(available_parts)+1):
    valid_choices.append(str(i))
print(valid_choices)
current_choice = "-"
computer_parts = []

while current_choice != '0':
    if current_choice in valid_choices: 
        index=int(current_choice)-1
        chosen_part = available_parts[index]
        if chosen_part in computer_parts:
            #if the item is already in the list, so remove it
            print("Removing {}".format(current_choice))
            computer_parts.remove(chosen_part)
        else:
            print("Adding {}".format(current_choice))
            computer_parts.append(chosen_part)
        print(f"Your list now contains: {computer_parts}")
    else:
        print("Please add options from the list below:")
        for number,part in enumerate(available_parts):
            print("{0}: {1}".format(number + 1, part))

    current_choice = input()

print(computer_parts)
```

### Removing items from a list Backwards

```python
#Removing unwanted items from a unsorted list, can be done much more easier than removing unwanted items foreward.
```

```python
#We have seen that we cannot remove items from a list, while iterating over it foreward. We remove an item, the size of 
#the list changes, and will not work properly
```

```python
#Which basically messes up the index numbers as we move towards the list.
```

```python
#But if we were to remove an item from the list, while iterating over it backwards, all the items index after it will be
#changed. But the items indexes before it, will stay exactly the same.
```

```python
#Iterating backwards is a valuable techinque & allows the size of your sequence to be changed without causing a problem.
```

##### Lets see two ways to do that

```python
#Using for loop to iterate backwards
```

```python
data=[104,101,4,105,308,105,5,107,100,306,106,102,108]
min_valid=100
max_valid=200

for index in range(len(data)-1,-1,-1):
    if data[index] < min_valid or data[index] > max_valid:
        print(index,data)
        del data[index]

print(data)
```

```python
#Using Reserved Function
```

```python
#The advantage of using reversed is that it let us use enumerate
```

```python
data=[104,101,4,105,308,105,5,107,100,306,106,102,108]
min_valid=100
max_valid=200

top_index=len(data)-1 #The reason we are using this because we are displaying the index position of the item we are
                      #deleting, you can ignore this, if you don't want to display all index values and just want the
                      #final list.
for index,value in enumerate(reversed(data)):
    if value<min_valid or value>max_valid:
        print(top_index - index,value)
        del data[top_index - index]

print(data)
```

```python
#The 'end' argument
```

```python
#for end to see in action, we will use the code from the previous section
```

```python
menu=[
    ["eggs","bacon"],
    ["eggs","sausage","bacon"],
    ["eggs","spam"],
    ["eggs","bacon","spam"],
    ["eggs","bacon","sausage","spam"],
    ["spam","bacon","sausage","spam"],
    ["spam","eggs","spam","spam","bacon","spam"],
    ["spam","bacon","spam","sausage","spam","tomato"],
]

for meal in menu:
    for item in meal:
        if item !="spam":
            print(item, end=" ") #We have added end argument here
    print()
```

```python
# What this end argument does, instead of printing a new line after each item, they will now be seperated by " "
```

```python
#The print() in line 16 is needed, otherwise every item will be in the same line
```

```python
# Alright. Now we can't use sep here,
# because we've only passed one object to print - you can see that on line 15.
# Remember that the separator's only used between the values we print, when we pass
# more than one value to the print function. However, we can still get a
# comma between the meal items, by putting a comma in the string that we used for
# end, so let's do that.
```

```python
for meal in menu:
    for item in meal:
        if item !="spam":
            print(item, end=", ") #We have added end argument here
    print()
```

```python
# That's not bad, but did you expect that extra comma at the end of each line? It's not
# actually easy to remove that, with this code. In fact, the most pythonic way to
# get that output without a trailing comma, is to use a generator expression.
```

```python
# (Now generator expressions are advanced Python expression, we will learn about it later)
```

```python
# Let's see the generator code, that will remove the comma from the end
```

```python
for meal in menu:
    items = ', '.join((item for item in meal if item != "spam"))
    print(items)
```
