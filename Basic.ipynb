```python
#Operator Precedence: Follows the same rule as BODMAS
```

```python
a=12
b=3
```

```python
print(a+b/3-4*12)
```

```python
print(a+(b/3)-(4*12))
```

```python
#If we want 12 as the answer.
```

```python
print((((a+b)/3)-4)*12)
print(((a+b)/3-4)*12) #as division is higher precedence than subtraction we can avoid a bracket.
```

```python
c=a+b
d=c/3
e=d-4
print(e*12)
```

```python
#NOTE:The problem with operator precednce is that many people believe that divison have higher precedence than multiplication and addition have higher precedence than sutraction
```

```python
#Thats not the case, division and multiplication havae equal precedence and addition and subtraction have equal precedence
```

```python
#In an expression that mixes operation with equal precedence, they're evaluted from left to right
```
